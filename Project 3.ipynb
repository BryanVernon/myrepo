```python
import matplotlib.pyplot as plt
plt.hist( NBA['AGE'] , bins = 40)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of age')
```

#### This graph shows the distribution of age amoung regular season NBA players in the 2019-2020 season. We see that the data is relatively normal with the exception of a handful of players that are abnormally older. These are the players with really long contracts that the team keeps to keep bringing fans to the games.

```python
plt.bar(NBA['POS'],NBA['PPG'])
plt.xlabel('Position')
plt.ylabel('PPG')
plt.title('Average Points per Game based on position')
```

#### This is a visulization of the categorical variable of position related to points per game. As we can see, the Guards score the most points on average and this is expected. It is also normal for the centers to have relatively lower points per game.
